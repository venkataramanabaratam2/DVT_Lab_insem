```python
import matplotlib.pyplot as plt

# Basketball players data
basketball_players = ["Player A", "Player B", "Player C", "Player D"]
basketball_points = [25, 30, 20, 18]

# grouped bar chart for basketball players' points scored
plt.figure(figsize=(10, 5))
plt.bar(basketball_players, basketball_points, color='b', alpha=0.7)
plt.xlabel('Basketball Players')
plt.ylabel('Points Scored')
plt.title('Basketball Players\' Points Scored')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Football players data
football_players = ["Player 1", "Player 2", "Player 3", "Player 4"]
football_goals = [15, 10, 20, 12]

# grouped bar chart for football players' goals scored
plt.figure(figsize=(10, 5))
plt.bar(football_players, football_goals, color='yellow', alpha=0.7)
plt.xlabel('Football Players')
plt.ylabel('Goals Scored')
plt.title('Football Players\' Goals Scored')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assists data
assists_players = ["Player A", "Player B", "Player C", "Player D", "Player 1", "Player 2", "Player 3", "Player 4"]
assists_values = [10, 8, 12, 15, 7, 10, 5, 8]

# horizontal bar chart to compare assists
plt.figure(figsize=(10, 5))
plt.barh(assists_players, assists_values, color=['b', 'b', 'b', 'b', 'yellow', 'yellow', 'yellow', 'yellow'])
plt.xlabel('Assists')
plt.ylabel('Players')
plt.title('Comparison of Assists: Basketball vs. Football Players')
plt.show()
```
